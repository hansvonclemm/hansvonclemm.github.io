```python
import pandas as pd
import datetime
import os
from glob import glob
```

```python
# some functions from past calculations
def clean_datetime(df):
    pat = '(?P<month>\d{2})/(?P<day>\d{2})  (?P<hour>\d{2}):(?P<minute>\d{2}):(?P<second>\d{2})'
    exp = df['Date/Time'].str.extract(pat,expand=True)
    exp['hour'] = exp['hour'].replace(24,0)
    exp['year'] = 2014

    df = df.set_index(pd.to_datetime(exp))

    return df  

```

```python
# we'll do this math for a HIGH load home in Augusta, Maine. 

sampleData_augusta = pd.read_csv('./data/USA_ME_Augusta.AP.726185_TMY3_HIGH.csv')
augusta = clean_datetime(sampleData_augusta)
augusta

# now I want to resample this by days starting at 12NOON -

# also resample 
# then you can throw out and days where it's not that cold at night. 
```
